## Task 6

#### Make the series Stationary and use the relevant model for the Time Series. Store your fitted model variable named as ts_model. Give reasons for choosing your option. (MM - 25)

```python
# Computing the log
birth_data_log = np.log(birth_data.DailyCount)

# Decomposing the data to get the trend, seasonality and the residuals
decomposition = seasonal_decompose(np.asarray(birth_data_log), freq=3)
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid


fig=plt.figure(figsize=(15,10))

# Original graph
plt.subplot(411)
plt.plot(birth_data_log, label='Original')

# Trend graph
plt.subplot(412)
plt.plot(trend, label='Trend')

# Seasonality graph
plt.subplot(413)
plt.plot(seasonal,label='Seasonality')

# Residual graph
plt.subplot(414)
plt.plot(residual, label='Residual')

plt.legend(loc='best')
plt.show()
```

### Ans - Observations

We can see from above decomposition that the **Trend Graph almost traces the Original graph** that adds the validity to assumption we made of slightly an increasing trend. Also the **Seasonality Graph verify that there is no seasonal trend** because it is throughout constant. So to get the fittest model for the data given first we need to detrend the graph.

```python
# Getting the log difference of current and log and shifted log
birth_data_log_diff = birth_data_log - birth_data_log.shift()
```

```python
# Again computing the log on the different value we obtained
dynamic_mean = birth_data_log_diff.rolling(window=12,center=False).mean()     
dynamic_var =birth_data_log_diff.rolling(window=12,center=False).var() 

fig = plt.figure(figsize=(15,6))

original = plt.plot(birth_data_log_diff,label='Original')
mean = plt.plot(dynamic_mean, color='red', label='Dynamic Mean')
var = plt.plot(dynamic_var, color='green', label = 'Dynamic Var')

plt.legend(loc='best') 
plt.title('Dynamic Mean & Variance')
plt.show(block=False)
```

### Ans - Observations

Now since we de-trend the data with log differnce we can see that the Dynamic Mean and Dynamic Variation lies between small value of y. Hence the time series generated is now a **Stationary Time Series**. 

```python
fig = plt.figure(figsize=(13,10))

# Plotting the ACF Plot with lags of 40 and with squeezed values so the data get adjusted within the limits 
# to get the perfect model
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(birth_data_log.values.squeeze(), lags=40, ax=ax1)

# Plotting PACF plot
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(birth_data_log, lags=40, ax=ax2)

plt.show()
```

### Ans - Observations

We can observe that the plot of Autocorrelation is not getting any proper cutoff after the nth log term which is required for it to be MA series,**Hence it is not a MA Series**. From the Pacf Plot we can see that the graph gets cutoff after the 2nd term hence it states that **it is a AR Series with order of 2**.

## Creating Models

### ARMA Model

```python

# Creating the ARMA model with the parameters of AR And MA series 2,0 as p,q
ts_model_arma = sm.tsa.ARMA(birth_data_log, (2,0)).fit(disp=False)

figure = plt.figure(figsize=(20,8))
plt.plot(birth_data_log)

# Plotting the graph for the fitted values of the model we created
plt.plot(ts_model_arma.fittedvalues,color='red')
plt.title("ARMA Model")
plt.show()
```
